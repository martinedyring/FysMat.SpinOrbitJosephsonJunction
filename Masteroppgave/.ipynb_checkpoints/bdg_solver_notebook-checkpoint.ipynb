```python
%%cython --annotate
cimport cython
import numpy as np
cimport numpy as np
from cython.parallel import prange
cdef extern from "complex.h":
    double complex cexp(double complex)
cdef extern from "math.h":
    double sin(double x) nogil
    double cos(double x) nogil


#↑↓ 
# F_i↑↓x+
# F_i↑↓x-
cdef int INDEX_F_I_UPDOWN = 0
cdef int INDEX_F_I_UPDOWN_XP = 1
cdef int INDEX_F_I_UPDOWN_XM = 2
cdef int INDEX_F_I_UPDOWN_YP = 3
cdef int INDEX_F_I_UPDOWN_YM = 4
cdef int INDEX_F_I_UPUP_XP = 5
cdef int INDEX_F_I_DOWNDOWN_XP = 6
cdef int F_DIMENSION_2 = 7
    
@cython.boundscheck(False)
cdef double complex U_term(int i, double U, double complex [:,:] F) nogil:
    return -U*F[i, INDEX_F_I_UPDOWN]

@cython.boundscheck(False)
cdef double complex hamilton_epsilon(int i, int j, float k, int spini, double [:,:] t_ijs_x, double[:,:] t_is_y, double [:,:] h_i) nogil:
    """
    Calculates the diagonal epsilon_ijk as defined elsewere.
    
    Parameters
    ----------
    i: int
        site i
    j: int
        site j
    k: float
        wavevector in y_direction
    spini: int
        spin at site i
    spinj: int
        spin at site j
    """
    cdef double h = 0.0
    cdef double ty = 0.0
    cdef double tx = 0.0
    if i == j:
        h = h_i[i,2] # 2 gives z component
        ty = t_is_y[i, spini-1] # spini in (1, 2) => (0, 1) index => (spinup, spindown)
    elif i == j+1:
        tx = t_ijs_x[j, spini-1]
    elif i == j-1:
        tx = t_ijs_x[i, spini-1]
    if spini == 2:
        h = -h
    if i == j+1 or i == j-1:
        diffij = 1.0
    cdef double complex e = -(h + 2*ty*cos(k)) + tx
    return e
    

@cython.boundscheck(False) # turn off bounds-checking for entire function
#@cython.wraparound(False)  # turn off negative index wrapping for entire function
cdef void set_diagonal(double complex * arr, 
                       int N_arr, 
                       int i, 
                       int j, 
                       float k, 
                       float mu,  
                       double [:,:] t_ijs_x, 
                       double[:,:] t_is_y, 
                       double [:,:] h_i) nogil:
    """
    
    Returns
    -------
    res: np.ndarray[np.complex64, ndim=2]
    """
    arr[0] = hamilton_epsilon(i, j, k, 1, t_ijs_x, t_is_y, h_i)-mu
    arr[N_arr + 1] = hamilton_epsilon(i, j, k, 2, t_ijs_x, t_is_y, h_i)-mu
    arr[2*(N_arr+1)]= -hamilton_epsilon(i, j, k, 1, t_ijs_x, t_is_y, h_i)+mu
    arr[3*(N_arr+1)] = -hamilton_epsilon(i, j, k, 2, t_ijs_x, t_is_y, h_i)+mu

@cython.boundscheck(False) # turn off bounds-checking for entire function
#@cython.wraparound(False)  # turn off negative index wrapping for entire function
@cython.cdivision(True)
cdef void set_hubbard_u(double complex * arr, int N_arr, int i, int j, float k, double U, double complex [:,:] F) nogil:
    if i == j:
        arr[3] = U_term(i, U, F)/2
        arr[N_arr + 2] = -U_term(i, U, F)/2
        arr[2 * N_arr + 1] = -U_term(i, U, F).conjugate()/2
        arr[3 * N_arr] = U_term(i, U, F).conjugate()/2

@cython.boundscheck(False)
cdef void set_hxhy(double complex * arr, int N_arr, int i, int j, double [:,:] h_i) nogil:
    if i == j:
        arr[1] = h_i[i, 0] - 1.0j*h_i[i, 1] # (h_iik)_12 = h^x -ih^y
        arr[N_arr] = h_i[i, 0] + 1.0j*h_i[i, 1] # (h_iik)_21 = h^x + ih^y
        arr[2*N_arr + 3] = -h_i[i, 0] - 1.0j*h_i[i, 1] # (h_iik)_34 = -h^x -ih^y
        arr[3*N_arr + 2] = -h_i[i, 0] + 1.0j*h_i[i, 1] # (h_iik)_43 = -h^x +ih^y

@cython.boundscheck(False) # turn off bounds-checking for entire function
@cython.wraparound(False)  # turn off negative index wrapping for entire function
cpdef np.ndarray[np.complex128_t, ndim=2] get_ham(int N_x, 
                                                  float k, 
                                                  float mu, 
                                                  double complex[:,:] F,
                                                  double U,
                                                  double [:,:] t_ijs_x, 
                                                  double[:,:] t_is_y, 
                                                  double [:,:] h_i):
    cdef double complex [:,:] res = np.zeros((4*N_x, 4*N_x), dtype=np.complex128)
    cdef int i, j
    for i in range(N_x):
        for j in prange(N_x, nogil=True):
            set_diagonal(&res[4*i, 4*j], 4 * N_x, i, j, k, mu, t_ijs_x, t_is_y, h_i)
            set_hubbard_u(&res[4*i, 4*j], 4*N_x, i, j, k, U, F)
            set_hxhy(&res[4*i, 4*j], 4*N_x, i, j, h_i)
    _res = np.asarray(res)
    return _res

def pget_ham(N_x, 
             k, 
             mu, 
             F,
             U,
             t_ijs_x, 
             t_is_y, 
             h_i):
    assert F.shape == (N_x, F_DIMENSION_2), "F has not the shape of (%i, %i). Got %s" % (N_x, F_DIMENSION_2, str(F.shape))
    assert t_ijs_x.shape == (N_x-1, 2), "t_ijs_x has not the shape of (%i, %i). Got %s" % (N_x-1, F_DIMENSION_2, str(t_ijs_x.shape))
    assert t_is_y.shape == (N_x, 2), "t_is_y has not the shape of (%i, %i). Got %s" % (N_x, 2, str(t_is_y.shape))
    assert h_i.shape == (N_x, 3), "h_i has not the shape of (%i, %i). Got %s" % (N_x, 3, str(h_i.shape))
    return get_ham(N_x, k, mu, F, U, t_ijs_x, t_is_y, h_i)
```

```python
def check_hermiticity(op):
    return np.allclose(np.zeros_like(op), op - (op.T).conjugate())
```

```python
N_X = 200
N_SC = 100
N_F = 50
F = np.ones((N_X, F_DIMENSION_2), dtype=np.complex128)*1.0j
t_sc = 1.3
t_f = 1.59
t_sc_f = 0.678
h_f = 0.333
U = 1.0

t_ijs_x = np.zeros((N_X-1, 2), dtype=np.float64)
t_is_y = np.zeros((N_X, 2), dtype=np.float64)
h_i = np.zeros((N_X, 3), dtype=np.float64)
for i in range(N_X):
    if i < N_SC:
        t_is_y[i, 0] = t_sc
        t_is_y[i, 1] = t_sc
    else:
        t_is_y[i, 0] = t_f
        t_is_y[i, 1] = t_f
        h_i[i, 2] = h_f
for i in range(N_X-1):
    if i < N_SC-1:
        t_ijs_x[i,0] = t_sc
        t_ijs_x[i,1] = t_sc
    elif i == N_SC-1:
        t_ijs_x[i,0] = t_sc_f
        t_ijs_x[i,1] = t_sc_f
    else:
        t_ijs_x[i,0] = t_f
        t_ijs_x[i,1] = t_f    
```

```python
h = pget_ham(N_X, 0.0, 0.0, F, U, t_ijs_x, t_is_y, h_i)
assert check_hermiticity(h)
```
