```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(value / (np.pi/4)))
    if N == 0:
        return "0"
    elif N == 1:
        return  r"$\pi/4$"
    elif N == 2:
        return r"$\pi/2$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)



x = theta_array
y = free_energy_array
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xticks([0, np.pi/4, np.pi/2])
ax.set_xticklabels(['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$'])
#ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
#ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/3))
#ax.set_xticklabels([r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$'])
ax.plot(x, y-max(y), color='limegreen')
ax.plot(x, y-max(y), 'o', color='limegreen')
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$F_N - max(F_{N,n_y=0})$')
#plt.rc('font', size=100)
#plt.rc('axes', labelsize=100)  
#plt.rc('xtick', labelsize=50)    # fontsize of the tick labels
plt.rc('ytick', labelsize=50)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.xaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontsize(20)
plt.savefig('free_energy_lambda_angle_xz, stength=0.8_final_label.png', dpi=200, bbox_inches='tight')
plt.show()
```
