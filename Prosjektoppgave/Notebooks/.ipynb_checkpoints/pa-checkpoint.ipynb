```python
N_x = 2
res = np.zeros((4*N_x, 4*N_x),  dtype=np.complex128)

for i in range(4*N_x):
    for k in range(4*N_x):
        res[i][k] = (i+1)*(k+1) + k*1.0j
```

```python
print(res)
#print(np.real(res))
#print(np.conj(res))
```
