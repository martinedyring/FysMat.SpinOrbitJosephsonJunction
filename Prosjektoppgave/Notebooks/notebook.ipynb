```python
import numpy as np

#↑↓ 
# F_i↑↓x+
# F_i↑↓x-
INDEX_F_I_UPDOWN = 0
INDEX_F_I_UPDOWN_XP = 1
INDEX_F_I_UPDOWN_XM = 2
INDEX_F_I_UPDOWN_YP = 3
INDEX_F_I_UPDOWN_YM = 4
INDEX_F_I_UPUP_XP = 5
INDEX_F_I_DOWNDOWN_XP = 6
F_DIMENSION_2 = 7
F_DIMENSION_2py = F_DIMENSION_2

def U_term(i, U, F):# cdef double complex U_term(int i, double U, double complex [:,:] F) nogil:
    return -U*F[i, INDEX_F_I_UPDOWN]


def hamilton_epsilon(i, j, k, spini, t_ijs_x, t_is_y, h_i):#cdef double complex hamilton_epsilon(int i, int j, float k, int spini, double [:,:] t_ijs_x, double[:,:] t_is_y, double [:,:] h_i) nogil:
    """
    Calculates the diagonal epsilon_ijk as defined elsewere.
    
    Parameters
    ----------
    i: int
        site i
    j: int
        site j
    k: float
        wavevector in y_direction
    spini: int
        spin at site i
    spinj: int
        spin at site j
    """
    h = 0.0
    ty = 0.0
    tx = 0.0
    if i == j:
        h = h_i[i,2] # 2 gives z component
        ty = t_is_y[i, spini-1] # spini in (1, 2) => (0, 1) index => (spinup, spindown)
    elif i == j+1:
        tx = t_ijs_x[j, spini-1]
    elif i == j-1:
        tx = t_ijs_x[i, spini-1]
    if spini == 2:
        h = -h
    if i == j+1 or i == j-1:
        diffij = 1.0
    e = np.complex128(-(h + 2*ty*np.cos(k)) - tx)
    return e
    
    

def set_diagonal(arr, N_arr, i, j, k, mu, t_ijs_x, t_is_y, h_i): #cdef void set_diagonal(double complex * arr,int N_arr, int i, int j, float k, float mu, double [:,:] t_ijs_x, double[:,:] t_is_y, double [:,:] h_i) nogil:
    """
    
    Returns
    -------
    res: np.ndarray[np.complex64, ndim=2]
    """
    arr[0] = hamilton_epsilon(i, j, k, 1, t_ijs_x, t_is_y, h_i)-mu
    arr[N_arr + 1] = hamilton_epsilon(i, j, k, 2, t_ijs_x, t_is_y, h_i)-mu
    arr[2*(N_arr+1)]= -hamilton_epsilon(i, j, k, 1, t_ijs_x, t_is_y, h_i)+mu
    arr[3*(N_arr+1)] = -hamilton_epsilon(i, j, k, 2, t_ijs_x, t_is_y, h_i)+mu

def set_hubbard_u(arr, N_arr, i, j, k, U, F):#cdef void set_hubbard_u(double complex * arr, int N_arr, int i, int j, float k, double U, double complex [:,:] F) nogil:
    if i == j:
        arr[3] = U_term(i, U[i], F)/2
        arr[N_arr + 2] = -U_term(i, U[i], F)/2
        arr[2 * N_arr + 1] = -U_term(i, U[i], F).conjugate()/2
        arr[3 * N_arr] = U_term(i, U[i], F).conjugate()/2
        
def set_hxhy(arr, N_arr, i, j, h_i): #cdef void set_hxhy(double complex * arr, int N_arr, int i, int j, double [:,:] h_i) nogil:
    if i == j:
        arr[1] = h_i[i, 0] - 1.0j*h_i[i, 1] # (h_iik)_12 = h^x -ih^y
        arr[N_arr] = h_i[i, 0] + 1.0j*h_i[i, 1] # (h_iik)_21 = h^x + ih^y
        arr[2*N_arr + 3] = -h_i[i, 0] - 1.0j*h_i[i, 1] # (h_iik)_34 = -h^x -ih^y
        arr[3*N_arr + 2] = -h_i[i, 0] + 1.0j*h_i[i, 1] # (h_iik)_43 = -h^x +ih^y

def get_ham(N_x, k, mu, F, U, t_ijs_x, t_is_y, h_i): #cpdef np.ndarray[np.complex128_t, ndim=2] get_ham(int N_x, float k, float mu, double complex[:,:] F, double U, double [:,:] t_ijs_x, double[:,:] t_is_y, double [:,:] h_i):
    res = np.zeros((4*N_x, 4*N_x), dtype=np.complex128)
    for i in range(N_x):
        for j in range(N_x):
            set_diagonal(res, 4*N_x, i, j, k, mu, t_ijs_x, t_is_y, h_i)
            set_hubbard_u(res, 4*N_x, i, j, k, U, F)
            set_hxhy(res, 4*N_x, i, j, h_i)
            #set_diagonal(res[4*i:, 4*j:], 4 * N_x, i, j, k, mu, t_ijs_x, t_is_y, h_i)
            #set_hubbard_u(res[4*i:, 4*j:], 4*N_x, i, j, k, U, F)
            #set_hxhy(res[4*i:, 4*j:], 4*N_x, i, j, h_i)
    _res = np.asarray(res)
    return _res
def pget_ham(N_x,  k, mu, F, U, t_ijs_x, t_is_y, h_i):
    assert F.shape == (N_x, F_DIMENSION_2), "F has not the shape of (%i, %i). Got %s" % (N_x, F_DIMENSION_2, str(F.shape))
    assert t_ijs_x.shape == (N_x-1, 2), "t_ijs_x has not the shape of (%i, %i). Got %s" % (N_x-1, F_DIMENSION_2, str(t_ijs_x.shape))
    assert t_is_y.shape == (N_x, 2), "t_is_y has not the shape of (%i, %i). Got %s" % (N_x, 2, str(t_is_y.shape))
    assert h_i.shape == (N_x, 3), "h_i has not the shape of (%i, %i). Got %s" % (N_x, 3, str(h_i.shape))
    assert U.shape == (N_x,), "U_i has not the shape of (%i,). Got %s" % (N_x, str(U.shape))
    return get_ham(N_x, k, mu, F, U, t_ijs_x, t_is_y, h_i)
```

```python
def check_hermiticity(op):
    return np.allclose(np.zeros_like(op), op - (op.T).conjugate())
```

```python
N_X = 200
N_SC = 100
N_F = 50
u_sc = 0.0
u_f = 0.0
F = np.ones((N_X, F_DIMENSION_2py), dtype=np.complex128)*1.0j
t_sc = 1.0
t_f = 1.0
t_sc_f = 1.0
h_f = 0.0
U = np.zeros((N_X,), dtype=np.float64)

t_ijs_x = np.zeros((N_X-1, 2), dtype=np.float64)
t_is_y = np.zeros((N_X, 2), dtype=np.float64)
h_i = np.zeros((N_X, 3), dtype=np.float64)
for i in range(N_X):
    if i < N_SC:
        t_is_y[i, 0] = t_sc
        t_is_y[i, 1] = t_sc
        U[i] = u_sc
    else:
        t_is_y[i, 0] = t_f
        t_is_y[i, 1] = t_f
        h_i[i, 2] = h_f
        U[i] = u_f
for i in range(N_X-1):
    if i < N_SC-1:
        t_ijs_x[i,0] = t_sc
        t_ijs_x[i,1] = t_sc
    elif i == N_SC-1:
        t_ijs_x[i,0] = t_sc_f
        t_ijs_x[i,1] = t_sc_f
    else:
        t_ijs_x[i,0] = t_f
        t_ijs_x[i,1] = t_f    

h = pget_ham(N_X, np.pi/2, 0.0, F, U, t_ijs_x, t_is_y, h_i)
assert check_hermiticity(h)
vals, vecs = eigh(h, turbo=True)
plt.plot(vals)
```

```python
def plot_wave(w, n=0, channel='v', ax=None, x=None):
    bgn = 0
    if channel=='v':
        bgn = 1
    elif channel=='w':
        bgn = 2
    elif channel=='x':
        bgn = 3
    else:
        assert channel == 'u', "Parameter channel must be one of u, v, w, x. Got %s" % str(channel)
    wave = w[bgn::4, n]
    
    if ax is None:
        fig, ax = plt.subplots(1,1)
    if x is None:
        x = np.arange(0, wave.shape[0], 1)+1
    else:
        assert x.shape[0] == wave.shape[0]
    ax.plot(np.real(wave), label='real part')
    ax.plot(np.imag(wave), label='imag part')
    ax.set_xlim(min(x), max(x))
```
